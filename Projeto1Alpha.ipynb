```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#Iniciando as posições e os momentos além de definir o número de passos
n = 100
x = np.zeros(n)
Px = np.zeros(n)

#definindo a posição inicial
x[0] = 1

#Definindo o K do chute
k = 0.7

#calculando a posição inicial
Px[0] = k*math.sin(x[0]) 

#Iniciando a integração das equações de movimento
for i in range(0,n-1): 
    Px[i + 1] = Px[i] + k*math.sin(x[i])
    x[i + 1] = x[ i ] + Px[i + 1]
plt.plot(x,Px)
plt.show()
```
