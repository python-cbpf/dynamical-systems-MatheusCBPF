```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

```python
#Iniciando as posies e os momentos alm de definir o nmero de passos
n = 100
x = np.zeros(n)
y = np.zeros(n)
z = np.zeros(n)
#Definindo as posies iniciais
x[0] = np.random.random()
y[0] = np.random.random()
z[0] = np.random.random()
alpha = 1.0
#Definindo o K do chute
k = 0.

#tentando ajustar para o grfico 3D
fig = plt.figure()
ax = Axes3D(fig)

#Iniciando Zbarra e Ybarra e Alpha
zbar = 0.5
ybar = 0.5

for i in range(0,n-1):
    ybar = y[i]*math.cos(alpha) - z[i]*math.sin(alpha)
    zbar = y[i]*math.sin(alpha) + z[i]*math.cos(alpha)
    x[i + 1] = x[i]*math.cos(k*zbar) - ybar*math.sin(k*zbar)
    y[i + 1] = x[i]*math.sin(k*zbar) + ybar*math.cos(k*zbar)
    z[i + 1] = zbar

plt.rcParams['figure.figsize'] = (12,9)
ax.scatter(x,y,z,s=0.5)
plt.show()
```
