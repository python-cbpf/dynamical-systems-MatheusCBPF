```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

```python
#Iniciando as posições e os momentos além de definir o número de passos
n = 3000
ivalue = 400
x = np.zeros((ivalue,n))
y = np.zeros((ivalue,n))
z = np.zeros((ivalue,n))
#Definindo a quantidade de valores iniciais


#Definindo as posições iniciais
for i in range(0,ivalue):
    x[i,0] = np.random.uniform(-1, 1)  
    y[i,0] = np.random.uniform(-1, 1)*math.sqrt(1 - x[i,0]**2)
    z[i,0] = math.sqrt(1 - x[i,0]**2 - y[i,0]**2)
#Definindo o K do chute e alpha
k = 1.8
alpha = 1.0
#tentando ajustar para o gráfico 3D
fig = plt.figure()
ax = Axes3D(fig)

for j in range(0,ivalue):
    for i in range(0,n-1):
        ybar = y[j,i]*math.cos(alpha) - z[j,i]*math.sin(alpha)
        zbar = y[j,i]*math.sin(alpha) + z[j,i]*math.cos(alpha)
        x[j,i + 1] = x[j,i]*math.cos(k*zbar) - ybar*math.sin(k*zbar)
        y[j,i + 1] = x[j,i]*math.sin(k*zbar) + ybar*math.cos(k*zbar)
        z[j,i + 1] = zbar

for j in range(ivalue):
    plt.rcParams['figure.figsize'] = (12,9)
    ax.scatter(x[j,:],y[j,:],z[j,:],s=0.2)

        
#plt.rcParams['figure.figsize'] = (12,9)
#ax.scatter(x,y,z,s=0.5)
#plt.show()
```
