```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#Iniciando as posições e os momentos além de definir o número de passos
n = 5000
x = np.zeros(n)
Px = np.zeros(n)

#Definindo o K do chute
k = 1.12

#Definindo quantidade de condições iniciais
ivalue = 30
VetorX = np.zeros((ivalue,n))
VetorP = np.zeros((ivalue,n))

#Definindo as posições iniciais
for i in range(0,ivalue):
    VetorX[i,0] = np.random.random()    #(i + 0.3)*0.75 

#Definindo os momentos iniciais
for i in range(0,ivalue):
    VetorP[i,0] = np.random.random()
    
#definindo a posição inicial
x[0] = 1


#calculando a posição inicial
Px[0] = k*math.sin(x[0]) 

#Iniciando a integração das equações de movimento
for j in range(0,ivalue):
    for i in range(0,n-1): 
        VetorP[j,i + 1] = VetorP[j,i] + k*math.sin(VetorX[j,i])
        VetorX[j,i + 1] = VetorX[j,i] + VetorP[j, i + 1]
        VetorX[j,:] = np.mod(VetorX[j,:],2.*np.pi)
        VetorP[j,:] = np.mod(VetorP[j,:],2.*np.pi)
```

```python
#plt.plot(VetorX,VetorP,'.')
#plt.plot(VetorX[0,:],VetorP[0,:],'.')
for j in range(ivalue):
    plt.rcParams['figure.figsize'] = (12,9)
    plt.scatter(VetorX[j,:],VetorP[j,:],s=0.6)
    
```
